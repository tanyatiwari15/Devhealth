```python
from ultralytics import YOLO
```

```python
results = model.predict(source='C:/Users/Dell/AlignAI/dataset/images/test', save=True)
```

```python
results
```

```python
import pandas as pd

output_data = []

for res in results:
    image_name = res.path
    for det in res.boxes:
        label = det.cls
        conf = det.conf
        bbox = det.xyxy  # Bounding box coordinates
        # Check if bbox has elements, and convert to list if it does
        bbox_values = bbox.tolist()[0] if bbox.numel() == 4 else [None, None, None, None]
        output_data.append([image_name, label, conf, *bbox_values])

# Convert to DataFrame
columns = ['Image', 'Label', 'Confidence', 'x_min', 'y_min', 'x_max', 'y_max']
df_results = pd.DataFrame(output_data, columns=columns)

# Display or save DataFrame
print(df_results)
df_results.to_csv('prediction_results.csv', index=False)
```

```python
import pandas as pd

# Load the CSV files
predictions_df = pd.read_csv("prediction_results.csv")
true_labels_df = pd.read_csv("dataset/shuffled_image_labels.csv")

# Extract image name from the path in the 'Image' column of predictions_df
predictions_df['Image'] = predictions_df['Image'].apply(lambda x: x.split('\\')[-1])

# Convert the tensor label to a numerical value
predictions_df['Label'] = predictions_df['Label'].apply(lambda x: int(float(x.strip("tensor([])"))))

# Merge the dataframes on the 'Image' column to match the images
merged_df = pd.merge(predictions_df, true_labels_df, on='Image', suffixes=('_pred', '_true'))

# Calculate incorrect predictions
incorrect_predictions = merged_df[merged_df['Label_pred'] != merged_df['Label_true']]
num_incorrect = len(incorrect_predictions)

# Calculate accuracy
accuracy = (len(merged_df) - num_incorrect) / len(merged_df) * 100

# Display the results
print(f"Number of incorrect predictions: {num_incorrect}")
print(f"Accuracy: {accuracy:.2f}%")
```
